```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/diabetes/diabetes_prediction_dataset.csv')
```

```python
df.describe()
```

```python
df=df.drop_duplicates().reset_index(drop=True)
df
```

```python
for i in df.columns:
  print('{} - {}'.format(i,len(df[i].unique())))
```

```python
X = df.drop('diabetes', axis=1)
y = df['diabetes']
```

```python
from sklearn.preprocessing import OneHotEncoder
columns_to_encode = ['gender', 'smoking_history']
X_categorical = df[columns_to_encode]
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoder.fit(X_categorical)
X_encoded = encoder.transform(X_categorical)
encoded_column_names = encoder.get_feature_names_out(columns_to_encode)
X_encoded_df = pd.DataFrame(X_encoded, columns=encoded_column_names)
X = pd.concat([X.drop(columns_to_encode, axis=1), X_encoded_df], axis=1)
X
```

```python
from sklearn.preprocessing import StandardScaler
columns_to_scale = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']
X_num = X[columns_to_scale]
scaler = StandardScaler()
scaler.fit(X_num)
X_scaled = scaler.transform(X_num)
X_scaled_df = pd.DataFrame(X_scaled, columns=columns_to_scale)
X = pd.concat([X.drop(columns_to_scale, axis=1), X_scaled_df], axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
age = float(input("Enter the patient's age: "))
gender = input("Enter the patient's gender (Male/Female/Other): ")
heart_disease = int(input("Does the patient have a heart disease? (0 for No, 1 for Yes): "))
hypertension = int(input("Does the patient have hypertension? (0 for No, 1 for Yes): "))
smoking_history = input("Enter the patient's smoking history: ")
bmi = float(input("Enter the patient's BMI: "))
HbA1c_level = float(input("Enter the patient's HbA1c level: "))
blood_glucose_level = float(input("Enter the patient's blood glucose level: "))
```

```python
input_dict = {
    'age': age,
    'gender': gender,
    'hypertension': hypertension,
    'heart_disease': heart_disease,
    'smoking_history': smoking_history,
    'bmi': bmi,
    'HbA1c_level': HbA1c_level,
    'blood_glucose_level': blood_glucose_level
}

input_df = pd.DataFrame([input_dict])
input_categorical = input_df[['gender', 'smoking_history']]
input_encoded = encoder.transform(input_categorical)
input_encoded_df = pd.DataFrame(input_encoded, columns=encoded_column_names)
input_processed = pd.concat([input_df.drop(['gender', 'smoking_history'], axis=1), input_encoded_df], axis=1)
input_numerical = input_processed[columns_to_scale]
input_scaled = scaler.transform(input_numerical)
input_scaled_df = pd.DataFrame(input_scaled, columns=columns_to_scale)
input_processed = pd.concat([input_processed.drop(columns_to_scale, axis=1), input_scaled_df], axis=1)
input_processed
```

```python
prediction=model.predict(input_processed)
```

```python
if prediction[0] == 1:
    print("The patient is predicted to be diabetic.")
else:
    print("The patient is predicted to be non-diabetic.")
```
